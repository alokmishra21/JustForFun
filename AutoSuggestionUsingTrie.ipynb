```python
class Node:
    def __init__(self, data):
        self.data = data
        self.childs = [None]*26
        self.isWord = False
        
class Trie:
    def __init__(self, node):
        self.root = node
        self.wordList = []
        
    def isLeaf(self, node):
        if len(node.childs) is 0:
            return True 
        return False
    
    def insert(self, s):
        i = 0
        temp = self.root
        while i < len(s) and temp != None:
            # search the data in the trie
            if temp.childs[ord(s[i].lower()) - 97] != None:
                temp = temp.childs[ord(s[i].lower()) - 97]
            else:
                # new node to be inserted
                node = Node(s[i])
                temp.childs[ord(s[i].lower()) - 97] = node
                temp = node
                
            i += 1
        temp.isWord = True;
                    
    def insertList(self, wordList):
        '''
            This function insert the word from
            the list of words
        '''
        for word in wordList:
            self.insert(word)
    
    def deepSearch(self, node, matchedWord):
        # found the leaf node
        if node.isWord:
            self.wordList.append(matchedWord + node.data)
        
        for x in range(26):
            if node.childs[x] != None:
                self.deepSearch(node.childs[x], matchedWord + node.data)
    
    def search(self, s):
        '''
            This function returns the list of
            auto suggestions
        '''
        i = 0
        temp = self.root;
        while i < len(s):
            flag = False
            for x in range(26):
                if temp.childs[x] != None and temp.childs[x].data == s[i]:
                    temp = temp.childs[x]
                    i += 1
                    flag = True
                    break
            # No search found
            if flag == False:
                break
                
        if i == len(s):
            if temp.isWord:
                self.wordList.append(s)
            # reached till the end of search pattern
            for x in range(26):
                if temp.childs[x] != None:
                    self.deepSearch(temp.childs[x], s)
            
        return self.wordList
        
    
        
node = Node(None)

trie = Trie(node)
trie.insertList(["do", "dog", "doggie", "awe", "awesome"])

wordToSearch = input("Enter the word to search ")
result = trie.search(wordToSearch)

print("\nsearch results: ")
for x in result:
    print(x)
    
if len(result) == 0:
    print("No result found!!!")
```
