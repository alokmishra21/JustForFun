```python
file1 = open('first.txt','r')
file2 = open('second.txt','r')
seq1 = []
seq2 = []
for line in file1:
    seq1.append(line)
for line in file2:
    seq2.append(line)
    
# Now DP for LCS 
lookup = []
index = []

def printMat(mat):
    for i in range(len(mat)):
        print(mat[i], end="")
        
printMat(seq1)
printMat(seq2)

def LCS(seq1, seq2):
    len1 = len(seq1)
    len2 = len(seq2)
    for i in range(len1+1):
        lookup.append([0]*(len2+1))
    for i in range(len1+1):
        for j in range(len2+1):
            if i == 0 or j == 0:
                lookup[i][j] = 0
            elif seq1[i-1] == seq2[j-1]:
                lookup[i][j] = 1 + lookup[i-1][j-1]
                index.append((i-1,j-1))
            else:
                lookup[i][j] = max( lookup[i-1][j], lookup[i][j-1] )
                
    for i in range(len(lookup)):
        print(lookup[i])
    print(lookup[i])
    print(index)
    print( "maximum LCS length : {}\n".format(lookup[len1][len2] ))
            
LCS(seq1,seq2)
    
def printSidebySide(seq1, seq2):
    len1 = len(seq1)
    len2 = len(seq2)
    totalLine = len1 + len2 - len(index)
    ind = 0
    matInd = index[ind]
    i = 0
    j = 0
    for line in range(totalLine):
        if i == matInd[0] and j == matInd[1]:
            print("\033[0;32;48m {0:<40}{1:>40}".format(seq1[i][:-1],seq2[j][:-1]))
            i += 1
            j += 1
            ind += 1
            if ind < len(index):
                matInd = index[ind]
        elif i < len1 and i < matInd[0]:
            print("\033[0;31;48m {0:<40}{1:>40}".format(seq1[i][:-1]," "))
            i += 1
        elif j < len2 and j < matInd[1]:
            print("\033[0;31;48m {0:<40}{1:>40}".format(" ",seq2[j][:-1]))
            j += 1
            
printSidebySide( seq1, seq2 )
    
```
